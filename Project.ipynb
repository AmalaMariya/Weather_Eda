1. WHAT IS THE AVERAGE TEMPERATURE IN EACH COUNTRY?

```python
import matplotlib.pyplot as plt
import seaborn as sns

avg_temp_country = df.groupby('country')['temperature_celsius'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=avg_temp_country.index, y=avg_temp_country.values, palette='coolwarm')
plt.title("Top 10 Countries by Average Temperature (Celsius)")
plt.xlabel("Country")
plt.ylabel("Average Temperature (°C)")
plt.xticks(rotation=45)
plt.show()
```

Conclusion:
The visualization highlights the top 10 countries with the highest average temperatures, providing a clear comparison of climate differences across these nations.

2. WHICH COUNTRIES HAVE THE HIGHEST AND LOWEST AIR QUALITY BASED ON PM2.5?

```python
avg_pm25_country = df.groupby('country')['air_quality_PM2.5'].mean().sort_values(ascending=False)

plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1)
sns.barplot(x=avg_pm25_country.head(5).index, y=avg_pm25_country.head(5).values, palette="Reds_r")
plt.title("Worst 5 Countries by PM2.5 Levels")
plt.xlabel("Country")
plt.ylabel("Average PM2.5")
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.barplot(x=avg_pm25_country.tail(5).index, y=avg_pm25_country.tail(5).values, palette="Greens")
plt.title("Best 5 Countries by PM2.5 Levels")
plt.xlabel("Country")
plt.ylabel("Average PM2.5")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

Conclusion:
The visualization highlights the countries with the worst and best average PM2.5 levels, showing a stark contrast between nations with poor air quality and those with cleaner air. It provides a clear comparison of global air quality extremes.

4. WHAT ARE THE DAYLIGHT HOURS IN EACH COUNTRY?

```python
# Convert sunrise and sunset to datetime and calculate daylight duration
df['sunrise'] = pd.to_datetime(df['sunrise'], format='%I:%M %p')
df['sunset'] = pd.to_datetime(df['sunset'], format='%I:%M %p')
df['daylight_hours'] = (df['sunset'] - df['sunrise']).dt.total_seconds() / 3600

avg_daylight_country = df.groupby('country')['daylight_hours'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=avg_daylight_country.index, y=avg_daylight_country.values, palette="YlOrBr")
plt.title("Top 10 Countries by Average Daylight Hours")
plt.xlabel("Country")
plt.ylabel("Average Daylight Hours")
plt.xticks(rotation=45)
plt.show()
```

Conclusion:The visualization displays the top 10 countries with the longest average daylight hours, illustrating global variations in daylight duration. It offers a clear comparison of how much daylight different countries experience on average.

11. WHAT IS THE DISTRIBUTION OF HUMIDITY LEVELS (LOW, MODERATE, HIGH) AMONG LOCATIONS?

```python
# Categorizing humidity
def categorize_humidity(humidity):
    if humidity < 30:
        return 'Low'
    elif 30 <= humidity < 60:
        return 'Moderate'
    else:
        return 'High'

df['humidity_category'] = df['humidity'].apply(categorize_humidity)
humidity_distribution = df['humidity_category'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(humidity_distribution, labels=humidity_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title("Humidity Levels Distribution")
plt.axis('equal')
plt.show()
```

Conclusion:The pie chart shows that the majority of humidity levels in the dataset fall within the "Moderate" range, with lesser occurrences of "Low" and "High" humidity.
